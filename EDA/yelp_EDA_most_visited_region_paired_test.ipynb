```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 특정 region_now를 정의
region_now = "New Orleans, 70130"

# 분석할 business_ids를 지정
business_ids = [147782, 26715, 3493, 18099, 24722]

# 각 business_id별로 분석 및 시각화
for business_id in business_ids:
    business_df = merged_df[merged_df['business_id'] == business_id]

    # visit_cnt가 0인 사람들의 stars 평균과 개수 계산
    zero_visit_stars = business_df[business_df['visit_cnt'] == 0]['stars']
    zero_visit_stars_mean = zero_visit_stars.mean()
    zero_visit_count = zero_visit_stars.count()

    # visit_cnt가 0이 아니면서 most_visited_region이 region_now인 사람들의 stars 평균과 개수 계산
    non_zero_visit_stars = business_df[(business_df['visit_cnt'] > 0) & (business_df['most_visited_region'] == region_now)]['stars']
    non_zero_visit_stars_mean = non_zero_visit_stars.mean()
    non_zero_visit_count = non_zero_visit_stars.count()

    # 결과 출력
    print(f"Business ID: {business_id}")
    print(f"Overall average stars: {business_df['stars'].mean()}")
    print(f"Average stars for users with visit_cnt = 0: {zero_visit_stars_mean} (count = {zero_visit_count})")
    print(f"Average stars for users with visit_cnt > 0 and most_visited_region = {region_now}: {non_zero_visit_stars_mean} (count = {non_zero_visit_count})")
    print("\n")

    # 히스토그램 데이터 생성
    bins = np.linspace(1, 5, 30)
    zero_visit_hist, bins = np.histogram(zero_visit_stars, bins=bins)
    non_zero_visit_hist, _ = np.histogram(non_zero_visit_stars, bins=bins)

    # 막대의 위치 설정
    bar_width = (bins[1] - bins[0]) / 2
    bar_positions = bins[:-1]  # 막대의 시작 위치

    # 그래프 생성
    plt.figure(figsize=(10, 6))

    plt.bar(bar_positions - bar_width/2, zero_visit_hist, width=bar_width, alpha=0.5, label=f'visit_cnt = 0 (n={zero_visit_count})')
    plt.bar(bar_positions + bar_width/2, non_zero_visit_hist, width=bar_width, alpha=0.5, label=f'visit_cnt > 0 and region = {region_now} (n={non_zero_visit_count})')

    plt.xlabel('Stars')
    plt.ylabel('Count')
    plt.title(f'Distribution of Stars for Business ID {business_id}')
    plt.legend()
    plt.show()
```

```python
import pandas as pd
import scipy.stats as stats
```

```python
# 동일 지역 여부를 나타내는 컬럼 추가
df['same_region'] = df['region'] == df['most_visited_region']

# business_id별로 그룹화
grouped = df.groupby('business_id')

# 결과를 저장할 리스트 초기화
results = []

# 각 그룹에 대해 동일 지역 유저와 비동일 지역 유저 비교
for business_id, group in grouped:
    same_region_users = group[group['same_region'] == True]
    different_region_users = group[group['same_region'] == False]

    # 그룹 내 평균 별점 계산
    same_region_avg_spend = same_region_users['stars'].mean() if not same_region_users.empty else 0
    different_region_avg_spend = different_region_users['stars'].mean() if not different_region_users.empty else 0

    # T-검정 수행
    if not same_region_users.empty and not different_region_users.empty:
        t_stat, p_value = stats.ttest_ind(same_region_users['stars'], different_region_users['stars'])
    else:
        t_stat, p_value = None, None  # 비교할 데이터가 없을 때

    # 결과 저장
    results.append({
        'business_id': business_id,
        'same_region_avg_spend': same_region_avg_spend,
        'different_region_avg_spend': different_region_avg_spend,
        't_stat': t_stat,
        'p_value': p_value
    })

# 결과를 데이터프레임으로 변환
result_df = pd.DataFrame(results)
print(result_df)
```

# 최빈 10개

```python
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# 가정: df라는 데이터프레임이 이미 주어졌다고 가정
# 예: df = pd.read_csv('your_data.csv')

# 1. 동일 지역 여부를 나타내는 컬럼 추가
df['same_region'] = df['region'] == df['most_visited_region']

# 2. business_id별 최빈값 상위 10개를 추출
top_10_business_ids = df['business_id'].value_counts().head(10).index

# 3. 최빈값에 해당하는 데이터만 필터링
filtered_df = df[df['business_id'].isin(top_10_business_ids)]

# 4. 각 business_id별 동일/비동일 지역 유저들의 평균 별점 및 t-test 계산
results = []

for business_id in top_10_business_ids:
    group = filtered_df[filtered_df['business_id'] == business_id]

    same_region_users = group[group['same_region'] == True]
    different_region_users = group[group['same_region'] == False]

    same_region_avg_stars = same_region_users['stars'].mean() if not same_region_users.empty else 0
    different_region_avg_stars = different_region_users['stars'].mean() if not different_region_users.empty else 0

    # T-검정 수행
    if not same_region_users.empty and not different_region_users.empty:
        t_stat, p_value = stats.ttest_ind(same_region_users['stars'], different_region_users['stars'], equal_var = False)
    else:
        t_stat, p_value = None, None

    results.append({
        'business_id': business_id,
        'same_region_avg_stars': same_region_avg_stars,
        'different_region_avg_stars': different_region_avg_stars,
        't_stat': t_stat,
        'p_value': p_value
    })

# 5. 결과를 데이터프레임으로 변환
result_df = pd.DataFrame(results)
print(result_df)

# 6. 최빈값 상위 10개의 결과를 2x5 레이아웃으로 시각화
plt.figure(figsize=(20, 10))  # 2x5 그리드를 위해 적절한 크기로 설정

for i, business_id in enumerate(top_10_business_ids, 1):
    plt.subplot(2, 5, i)
    sns.boxplot(x='same_region', y='stars', data=filtered_df[filtered_df['business_id'] == business_id],
                color='lightsteelblue')
    plt.title(f'business_id: {business_id}')
    plt.xlabel('Same Region')
    plt.ylabel('Stars')

plt.tight_layout()
plt.show()
```
